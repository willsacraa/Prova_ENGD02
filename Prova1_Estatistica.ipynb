```r
install.packages('readxl')
library(readxl)


AmeSul <- read_excel("America_do_Sul.xlsx")
brasil <- read_excel("BRASIL.xlsx")
View(brasil)
```

#Tabela descritiva

```r
media <- c(mean(brasil$new_cases),
           mean(brasil$new_deaths, na.rm = T),
           mean(brasil$reproduction_rate, na.rm = T))

mediana <- c(median(brasil$new_cases),
             median(brasil$new_deaths, na.rm = T),
             median(brasil$reproduction_rate, na.rm = T))

desvio <- c(sd(brasil$new_cases),
            sd(brasil$new_deaths, na.rm = T),
            sd(brasil$reproduction_rate, na.rm = T))

q1 <- c(quantile(brasil$new_cases, probs = 0.25),
        quantile(brasil$new_deaths, probs = 0.25, na.rm = T),
        quantile(brasil$reproduction_rate, probs = 0.25, na.rm = T))

q2 <- c(quantile(brasil$new_cases, probs = 0.50),
        quantile(brasil$new_deaths, probs = 0.50, na.rm = T),
        quantile(brasil$reproduction_rate, probs = 0.50, na.rm = T))

q3 <- c(quantile(brasil$new_cases, probs = 0.75),
        quantile(brasil$new_deaths, probs = 0.75, na.rm = T),
        quantile(brasil$reproduction_rate, probs = 0.75, na.rm = T))

q4 <- c(quantile(brasil$new_cases, probs = 1),
        quantile(brasil$new_deaths, probs = 1, na.rm = T),
        quantile(brasil$reproduction_rate, probs = 1, na.rm = T))

tabela <- data.frame(Media = media, Mediana = mediana, Desvio = desvio,
                     Quartil_1 = q1, Quartil_2 = q2, Quartil_3 = q3,
                     Quartil_4 = q4)

tabela <- round(tabela, 2)

rownames(tabela) <- c('Total Casos', 'Total Morte', 'Taxa de Reprodução')

View(tabela)
```

#Distribuicao

```r
H <- sum(brasil$reproduction_rate[(0:400)], na.rm = TRUE)
med <- mean(brasil$reproduction_rate[(0:400)], na.rm = TRUE)
des <- sd(brasil$reproduction_rate[(0:400)], na.rm = TRUE)

par(mfrow = c(1,1))

x <- rnorm(H, mean = med, sd = sqrt(des^2))
hist(x, freq = FALSE, xlab = 'Taxa de Reproducao do Covid19', ylab = 'Frequencia', 
     main = 'Distribuicao de Probabilidade Normal para Taxa de Reproducao do Covid19')
curve(dnorm(x,mean=med,sd=sqrt(des^2)),col=1,lty=1,lwd=1,add=TRUE) 
```
